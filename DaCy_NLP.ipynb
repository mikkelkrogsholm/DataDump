## Import DaCy and load large language model

```python
import dacy
nlp = dacy.load("large")
```

## Import httpx and get article from DR.dk

```python
import httpx

url = "https://www.dr.dk/nyheder/politik/hvad-mener-messerschmidt-egentlig-naar-han-kobler-mette-frederiksens-mink-fadaese"

r = httpx.get(url)

r
```

## Import BeatifulSoup and parse html to text

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, "html.parser")

article = soup.find("div", class_ = "dre-article-body").text

doc = nlp(article)
```

## Import pandas and create a sorted dataframe with counts of entities

```python
import pandas as pd

ents = []

for entity in doc.ents:
   print(entity, ":", entity.label_)
   ents.append(dict(entity = str(entity), label = entity.label_))

df_ents = pd.DataFrame(ents)

df_ents.groupby(['entity','label'])['entity'].size().reset_index(name='counts').sort_values("counts", ascending = False )
```
